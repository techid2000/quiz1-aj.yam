## Creating Plots

Jupyter notebooks are so cool because you can display plots created with code inline. The most popular library for plotting in library is matplotlib.

To get started with matplotlib, you need to import it:

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

It's also useful to import numpy for working with arrays:

```python
import numpy as np
```

Here's a basic example that graphs points along a sine wave:

```python
x = np.arange(0, 2*np.pi, 0.1) # create an array of numbers from 0 to 2pi, 0.1 interval
y = np.sin(x) # apply sin function to each x value

plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y)
```

### More plotting

Here's an example that shows the histograms of some random distributions:

```python
# plot uniform (blue), triangular (green), and almost normal (red) distributions
for times in [1, 2, 12]:
    vals = np.random.rand(1000)
    for i in range(1, times):
        vals += np.random.rand(1000)
    vals /= times
    plt.hist(vals)
```

### Even more plotting

You can even do 3D stuff. This is the graph of $z = \sin \sqrt{x^2 + y^2}$

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')

X = np.linspace(-10, 10, 300)
Y = np.linspace(-10, 10, 300)
X_grid, Y_grid = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X_grid**2 + Y_grid**2))
ax.plot_surface(X_grid, Y_grid, Z, cmap='rainbow')
ax.set_zlim(-2, 2)
plt.title('z = sin(sqrt(x^2 + y^2))')
```

## OpenCV with jupyter notebooks

Using OpenCV in jupyter notebooks is a great way to quickly prototype and experiment with computer vision. Because OpenCV seamlessly integrates with matplotlib and numpy, it's simple to use in Python:

```python
import cv2 # import opencv library

# load image as a numpy array
img = cv2.imread("img/flower.jpg", cv2.IMREAD_COLOR)
img.shape
```

The OpenCV Python bindings automatically load images as numpy arrays. The shape of the image loaded above is `(426, 640, 3)`, which means that the image is 640x426 and has 3 color channels.

To display the image, use `plt.imshow`. (Note that OpenCV loads images in BGR format. Matplotlib expects RGB images so conversion is needed. If `cvtColor` is too much typing, try `img[:,:,::-1]` to reverse the channels)

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

It's easy to manipulate images now. Here's an example of copying a patch of the image to another location:

```python
manipulated = img.copy()
manipulated[200:400,300:500,:] = manipulated[50:250,100:300,:]
plt.imshow(cv2.cvtColor(manipulated, cv2.COLOR_BGR2RGB))
```

Here's an example applying box blur to an image:

```python
kernel = np.ones((15, 15)) / 15**2 # create a 15x15 kernel whose elements are equal and sum to 1
blurred = cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
```

Here's how one could create a swirl effect. Use `alpha` to change the intensity and direction

```python
alpha = -0.018

y_axis, x_axis = np.arange(0, img.shape[0]), np.arange(0, img.shape[1])
# shift into coordinates where (0, 0) is center of image
y_axis, x_axis = y_axis - img.shape[0]/2, x_axis - img.shape[1]/2
grid_x, grid_y = np.meshgrid(x_axis, y_axis)
dist = np.sqrt(grid_x**2 + grid_y**2)
# don't swirl entire image
dist = np.clip(dist.max()/2 - dist, 0, 10000)
# apply rotation matrix where angle is dist*alpha
map_x = grid_x*np.cos(dist*alpha) - grid_y*np.sin(dist*alpha)
map_y = grid_x*np.sin(dist*alpha) + grid_y*np.cos(dist*alpha)

# shift back into opencv coordinates
map_y, map_x = map_y + img.shape[0]/2, map_x + img.shape[1]/2

swirl = cv2.remap(img, map_x.astype(np.float32), map_y.astype(np.float32), interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(swirl, cv2.COLOR_BGR2RGB))
```
